```python
import numpy as np
from numpy import random
```

```python
c=0.5
training_inputs=np.array([[2,0,1],[2,-2,1],[1,-2,-1]])
print("Training Inputs:\n",training_inputs)
```

```python
init_weights=random.rand(3)
print("Inititial Weights are:\n",init_weights)
```

```python
d=np.array([1,0,1])
print("D are:\n",d)
```

```python
for z in range(500):
    print("iterations:",z+1)
    j=0;
    for i in training_inputs:
        net=np.dot(i,init_weights.T)
        temp=c*(d[j]-net)
        del_w=np.dot(i,temp)
        init_weights+=del_w
        j=j+1
        print("Updated Weights are:",init_weights)
    print("\n")
    
```
